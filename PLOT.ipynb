```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from collections import Counter
import collections
%matplotlib inline
```

```python
data_origin = pd.read_csv('cleanedData.csv')
```

```python
data_origin
```

```python
def ID_cityfreq(ID_input):
    '''
    input:
        pokemonID
    return:
        the frequence in different cities(Series)
    '''
    data = data_origin.loc[data_origin['pokemonId'] == ID_input]
    city_freq_count = data['city'].value_counts()
    
    items= data['city'].tolist()
    letter_counts = Counter(items)
    df = pd.DataFrame.from_dict(letter_counts, orient='index')
    plt.hist(df, label ='frequency',color = 'navy')
    plt.savefig('Frequency of pokemon ' + str(ID_input) + 'appears in different cities')
    return city_freq_count
```

```python
ID_cityfreq(16)
```

```python
def city_IDfreq(city_input):
    ''''
    input:
        city name(str)
    return:
        the frequence of all pokemon appear in this city(Series)
    '''
    data = data_origin.loc[data_origin['city'] == city_input]
    ID_freq_count = data['pokemonId'].value_counts()
    ID_freq_count.hist()
    plt.savefig('Frequence of pokemons appear in ' + str(city_input))
    return ID_freq_count
```

```python
city_IDfreq('Mexico_City')
```

```python
def appeared_incity(ID_input, city_input):
    '''
    input:
        pokemonID(int) and city_name(str)
    
    return:
        **print** if the pokemon appeared in this city
    '''
    data = data_origin.loc[data_origin['pokemonId'] == ID_input]
    city_freq_count = data['city'].value_counts()
    if str(city_input) in city_freq_count.index:
        return (print('Pokemon ' + str(ID_input) + ' has appeared in ' + str(city_input)+ ' before'))
    else:
        return (print('Pokemon ' + str(ID_input) + ' has not appeared in ' + str(city_input)+ ' before'))
```

```python
appeared_incity(16, 'Mexico_City')
```

```python
def appeared_time(ID_input):
    '''
    input: 
        pokemonID(int) 
    return:
        pie chart of this pokemon appear in different time priod
        '''
    data = data_origin.loc[data_origin['pokemonId'] == ID_input]
    time_array = np.array(data['appearedTimeOfDay'])
    time_df = pd.DataFrame({'time':time_array})
    time_freq_count = time_df['time'].value_counts()
    
    #pie chart
    labels = 'night', 'morning', 'afternoon'
    pie(time_freq_count, labels =labels ,autopct='%1.1f%%')
    #save figure as png
    plt.savefig('pie chart of pokemon ' + str(ID_input) + ' showing up periods')
    return time_freq_count
```

```python
def co_occurance (ID_input):
    '''
    input:
        pokemonID(int)
    return:
        Five pokemonID(list)'''
    data = data_origin.loc[data_origin['pokemonId'] == ID_input]
    freq_list = []
    result = []
    for i in range(1,152):
        col_name = str('cooc_'+str(i))
        freq = sum(data[col_name])
        freq_list.append(freq)
    maxfive = sort(freq_list)[-5:]
    for i in range(1,6):
        a = freq_list.index(maxfive[-i])
        result.append(a)
        
    
    print(str(result) + 'occur ')
    return result
```
